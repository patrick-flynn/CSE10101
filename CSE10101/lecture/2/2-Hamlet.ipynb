Anyway, we should probably chop off all of the punctuation, and
also make all of the words lowercase. A handy pre-defined list of punctuation characters is described here: https://docs.python.org/2/library/string.html

```python
import string
string.punctuation
```

The following code runs over the list 'hamwords' and creates a new list 'hamwords2', where the items in hamwords2 have the punctuation marks removed, and all of the characters are made 'lowercase'.

```python
hamwords2 = []
for word in hamwords:
    outword = ''
    for c in word:
        if c not in string.punctuation:
            outword = outword + c
    hamwords2.append(outword.lower())
```

```python
# In[57]:

hamwords2[:150]
```

Now let's count up the words.

Use a _dictionary_ to store the word counts.

A dictionary can store any piece of data ("value"), using a "key" to locate and retrieve it.
Dicts are sometimes called "key-value stores".

In Python, we create dictionaries using curly brackets, and store/retrieve info using square brackets.

It's important to not mix up lists and dicts.


Here, we will store an integer (the value), for each unique word (the key).
The integer value will be the count of that word.

This might be called a "histogram" in some fields.

```python
counts = {}    # note: curly brackets mean "a new empty dictionary" here. 

for word in hamwords2:  # loop over the list
    if word not in counts:  # if it hasn't been used as a key
        counts[word] = 1 # create a value for this new key and initialize it to 1
    else:
        counts[word] = counts[word] + 1  # otherwise increment the value already there.
```

Remove entries associated with stopwords from the dictionary.  Deleting items from a dict uses an odd syntax.

To do this deletion, we have two steps.
1. iterate over the dictionary's keys, comparing each key to the stopword list. Remember the keys that are in the list.
2. iterate over the keys we remembered, deleting each one from the dictionary.

We can't delete from the dictionary on the fly because the
iterator expects "the thing being iterated over" to be fixed in size.

```python
todelete = []
for key in counts:
    if key in stopwords:
        todelete = todelete + [key]
```

```python
todelete # these should all be stopwords
```

```python
# now delete each of those keys from the dictionary (the count gets deleted too)

for key in todelete:
    print 'deleting ',key
    del counts[key]
```

So the entries in the dictionary are word counts, and the index used to access the dictionary is the word itself.

Obvious question: what word has the most uses?

A standard technique is to visit each entry and keep track of the largest count we've seen so far, and when we're done visiting, we should have the winner.

There is a Python idiom for iterating over the keys in a dictionary.

We can arrange the entries of the dictionary in descending-count order by
using a special
data structure called an OrderedDict, which is just like a dictionary,
but can be sorted.

The second line below creates the ordered dictionary. The dictionary is populated by the output of the __sorted__ function, and the __key__ basically tells you how to do the sort.  _The details of the sorting process are not important today._

```python
from collections import OrderedDict
sorted_counts = OrderedDict(sorted(counts.items(),key = lambda x:x[1],reverse=True))
sorted_counts
```
