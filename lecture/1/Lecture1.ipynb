##Get the data from the file##
A long-winded explanation of what's going on in the statement below.

Here is the statement: ``f = file('DJIA.csv','r')``

This statement "opens a file" so that we can read the information in it.  We do this by "calling a function" named __file__. A function is a little piece of code that can be used and reused. Some functions (like __file__) are supplied with the Python program; some can be added in to your Python environment by downloading from the Internet and installing them; and you can also write your own functions (we'll get to that later).

You can send info to functions. This function call has two parameters: the file name (in this case, DJIA.csv) and the "mode", which indicates whether you're reading the file, writing it, or both (we are reading, hence 'r').

Functions can return information to the caller. The information returned to the caller is called the "return value". It is the result of the function.

This statement is called an _assignment_ because Python will run the function on the right hand sde of the equal sign (passing it any parameter values you supply), and the return value will be stored in a _variable_ whose name appears on the left hand side of the equal sign.  So the result of the file() function is assigned to f.

The variable "f" gives us the ability to work with the file.  It is sometimes called a "file handle" or "file descriptor", and is a specific type of a more general class of item called an "object". You could think of an object as a little bundle of code and storage.  We can give f orders and read/write the data that is stored in the file that f refers to.

We are making progress. We know how to yank text out of a file, chop it into lines (each line is a string), split the lines into chunks (each chunk a string), and convert text strings to numbers.

What we need to do now is take the list of lines, and produce a list of numbers, which we can then plot.  This requires the ability to run the same code on each line, and collect up the results.  The simplest way to do this is to write a loop.

To start, we'll run some tests on a small subset of the list of lines (it's got 2610 lines!) so we don't have enormous output.

Recall that the lines are indexed: lines[0] is first, lines[1] is second, etc.  Here's a cute little trick. You can use Python's __slice__ syntax to get a subpart of a list.


```python
subset = lines[0:10]
```

```python
subset
```

Look carefully here. We got the first ten items in the file... ``lines[0]`` though ``lines[9]``.  Although the ``0::10`` syntax might suggest that we should get ``lines[0]`` through ``lines[10]``, we do not get the item indexed by the number on the right of the colon. You get everything up to it.

```python
subset2 = lines[1:3]
```

```python
subset2
```

Again, we get ``lines[1]`` and ``lines[2]``, but not ``lines[3]``.  Remember this or you will be beset by what we in the biz call "off-by-one" errors.

Back to our task. Let's get the subset again, but without the header line.

```python
subset = lines[1:10]; print subset
```

To separate the dates from the values, we can just use that splitter, right?

```python
splits = subset.split(',')
```

Nope.  the  __split__  function works on a string, not on a list.
Can we get each string out of the list, split it, and
assemble the results back into the list?

Yes.  This is an example of "iteration" or "looping" over the contents of the list.
There are several ways to do this. The approach immediately below is called a _list comprehension_, which may be ironic to the beginner.

```python
splits = [x.split(',') for x in subset]

splits
```

The code below does the same thing in a more traditional way.  The "for" syntax identifies a loop (the body of the loop is executed on each of the items in the subset list.

```python
splits = []
# loop over the items in subset
for s in subset:
    [date,value] = s.split(',')
    splits.append([date,value])
splits
```

```python
# We don't really need the date at all, because the data is sorted by date.
# So let's strip off the date part of the split, by selecting only the second
# item of each result

splits = [x.split(',')[1] for x in subset]
splits

```

```python
# Now get rid of the yucky \r\n on each line.
# Those characters are "carriage return, newline" and are often grouped
# together with the space ' ' and the tab character '\t' into a set called
# "whitespace". So we can strip the whitespace from the string using the
# cleverly named "strip" function:

splits = [x.split(',')[1].strip() for x in subset]
splits
```

```python
# FINALLY, let's convert the result from string to number.
#
splits = [float(x.split(',')[1].strip()) for x in subset]
splits
```

```python
# no quote character means these are numbers, not strings.
#
#plot this subset.  There is a nice plotting capability in the
# popular package "matplotlib".
#
# The "import" command causes Python to read in some code that provides
# new functions to Python.
#
# The %matplotlib thing is called a "magic" and we aren't going there right now.
#

import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(splits)
```

```python
# Spiffy!  But that's only nine data points. Let's crunch and plot all of the data!

splits = [float(x.split(',')[1].strip()) for x in lines]
```

```python
# NOOOOO What went wrong?
#
# We forgot: the first line of the file is the string "DATE,VALUE\r\n".
# You can't turn a character string into a number if it has non-numerical
# characters in it.
splits = [float(x.split(',')[1].strip()) for x in lines[1:]]
```

What happened this time? "could not convert string to float: ."
Debugging is hard, but there are some tried-and-true approaches.
One is to print debugging information as you go (can disable the printing later).
But we can't easily stick print functions into a list comprehension.

So expand it into a more verbose loop.


```python
splits = []
for s in lines[1:]:
    [date,value] = s.split(',')
    print 'the current value is ',value
    splits.append([float(value.strip())])
splits
```

So the problem is that some of the VALUEs are '.' rather than a number. Holidays, etc.

How to check for them and remove them?  Use a simple __conditional__ statement.

```python
splits = []
for s in lines[1:]:
    [date,value] = s.split(',')
    value = value.strip()
    if value == '.':
        print 'Discarding bogus value . for date: ',date
    else:
        splits.append([float(value.strip())])
        
splits
```

```python
plt.plot(splits)
```
