```python
fhamlet = file('Hamlet-pg2265.txt','r')
```

```python
hamtmp = fhamlet.readlines()
```

```python
# 381 determined via binary search.
# Note slice notation. If 2nd number not given, it means "to the end".
hamletlines = hamtmp[381:]
```

```python
len(hamletlines)
```

```python
hamletlines[0]
```

```python
# join the lines together and replace the \r\n with a space.
hamletstring = ' '.join(hamletlines).replace('\r\n',' ')
```

```python
hamletstring[:200]
```

```python
# we have one big long string containing all of Hamlet.
# let's break it into unique words.
# by default, split() will use 'whitespace' (space, tab, newline) as a divider.
hamwords = hamletstring.split()
```

```python
len(hamwords)
```

```python
hamwords[:150]
```

```python
# Holla.
# Ahead of his time: http://htmlgiant.com/random/lines-from-shakespeare-mistaken-for-1990s-hip-hop-lyrics/
#
#Anyway, we should probably chop off all of the punctuation, and
# also make all of the words lowercase.
# handy pre-defined list of punctuation characters is described here: https://docs.python.org/2/library/string.html

import string
string.punctuation
```

```python
hamwords2 = []
for word in hamwords:
    word2 = ''.join([c for c in word if c not in string.punctuation])
    hamwords2.append(word2.lower())
```

```python
hamwords2[:150]
```

```python
#
# Use a dictionary to gather up the word counts.
# a dictionary can store any piece of data under a supplied name.
# we will store an integer, for each entry whose name is a unique word in the set of words.
# the integer will maintain a count of the times that word has been seen.
#
# This might be called a "histogram" in some fields.
counts = {}    # note: curly brackets mean "dictionary"

for word in hamwords2:
    if word not in counts:
        counts[word] = 1
    else:
        counts[word] = counts[word] + 1
```

```python
counts
```

```python
counts['the']
```

```python
counts['holla']
```

```python
#
# let's get rid of extremely common English words, like 'the'.
#
fstop = file('stopwords.txt','r')
stopwords = fstop.readlines()
```

```python
stopwords = [s.strip() for s in stopwords]
```

```python
stopwords
```

```python
# remove entries associated with stopwords from the dictionary.
# to do this, we have two steps.
# 1. iterate over the dictionary's keys, comparing each
# key to the stopword list. Remember the keys that are in the list.
# 2. iterate over the keys, deleting each one from the dictionary.
#
# We can't delete from the dictionary on the fly because the
# iterator expects the thing being iterated over to be fixed in size.
todelete = []
for key in counts:
    if key in stopwords:
        todelete = todelete + [key]

for key in todelete:
    print 'deleting ',key
    del counts[key]
```

```python
len(counts)
```

```python
counts
```

```python
# Arrange te entries of the dictionary in descending order by
# the value (count) stored in the buckets addressed by the key. In order to do this, we have to use a special
# data structure called an OrderedDict, which is just like a dictionary,
# but can be sorted.
from collections import OrderedDict
sorted_counts = OrderedDict(sorted(counts.items(),key = lambda x:x[1],reverse=True))
sorted_counts
```
